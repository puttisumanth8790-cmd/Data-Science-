## Statistical Functions 

```python
Sum2=np.array([10,20,30,40])
```

```python
Sum2.mean() # mean of the array 
```

```python
Sum2.sum() # sum of the array 
```

```python
Sum2.max() # maximum value of the array 
```

```python
Sum2.min() # minimum value of the array 
```

```python
Sum2.std() # stamdard deviation of the array
```

```python
np.median(Sum2) # median of the array 
```

## Array Indexing

```python
sum1[0] # acessing the first element of the array 
```

```python
sum1[-1] # acessing the last element of the array
```

## 2D indexing 

```python
sum3=np.array([[1,2,3,4],[5,6,7,8],[2,4,4,5]])
```

```python
sum3[1,1] # acessing elements of the 2D array
```

```python
sum3[1] # acessing entre row 
```

```python
sum3[:,1] # acessing entire column
```
