## Matplotlib

```python
import matplotlib.pyplot as plt # importing librari Pyplot for plotting groups
```

```python
x=[1,2,3,4,5]      # we are creating two list and visualizing them
y=[10,20,30,40,50]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]       # adding labels and tittle
y=[10,20,30,40,50]
plt.plot(x,y)
plt.title("sales Trend")
plt.xlabel("month")
plt.ylabel("sales")
plt.show()

```

```python
plt.plot(x,y,color="red") # changing the line color
```

```python
plt.plot(x,y,linestyle="--") # changing the line style 
```

```python
plt.plot(x,y,marker="*") # changing the maker style 
```

```python
x=[10,20,30,40,50]       #scatter plot 
y=[15,20,25,30,35]     
plt.scatter(x,y)
plt.title("height and weight")
plt.xlabel("height")
plt.ylabel("weight")
plt.show()
```

```python
student=["A","B","C","D","E"] # bar graph
Marks=[10,20,30,40,50]
plt.bar(student,Marks)
plt.title("student Marks")
plt.xlabel("student")
plt.ylabel("marks")
plt.show()
```

```python
plt.barh(student,Marks) # horizontal bar chart
```

```python
import numpy as np
import matplotlib.pyplot as plt
marks=np.random.randint(40,100,1000)
plt.hist(marks)
plt.show()
```

```python
subjects=["English","Telugu","Hindi"]
Students=[40,50,60]
plt.pie(Students, labels=subjects)
plt.show()
```

```python
plt.pie(Students,labels=subjects,autopct="%1.1f%%") # display percentage in pie chart
```

```python
marks=[23,4,3,445,56,65,57,676,85,23]
plt.boxplot(marks)
plt.show()
```

```python
months=[1,2,3,4,5]  # multiple lines
sales1=[10,15,20,25,30]
sales2=[8,14,18,28,35]
plt.plot(months,sales1,label="probut A")
plt.plot(months,sales2,label="product B")
plt.grid(True) # Grid
plt.figure(figsize=(8,5)) # figure size
plt.savefig("chart.png")
plt.show()
```
