## Seaborn 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df=pd.read_excel("D:/Sumanth.xlsx")
```

```python
print(df.columns)
```

```python
sns.lineplot(x="Attendance ",y="Marks",data=df)
plt.show() # line plot
```

```python
sns.scatterplot(x="Attendance ",y="Marks",data=df)
plt.show()  # sacatter plot
```

```python
sns.barplot(x="Departement",y="Marks",data=df)
plt.show()  # Bar plot
```

```python
sns.countplot(x="Departement",data=df)
plt.show()  # count plot
```

```python
sns.histplot(df["Marks"]) #Histogram
plt.show()
```

```python
sns.histplot(df["Marks"],kde=True) ## histogram with kde
plt.show()
```

```python
sns.boxplot(y="Marks",data=df) # box plot
plt.show()
```

```python
sns.boxplot(x="Departement",y="Marks",data=df) # Box plot by category
plt.show()
```

```python
sns.violinplot(x="Departement",y="Marks",data=df) # violin plot
plt.show()
```

```python
sns.pairplot(df) # pair plot
plt.show()
```

```python
corr=df.corr(numeric_only=True) # heat map
sns.heatmap(corr,annot=True)
plt.show()
```

```python
sns.scatterplot(x="Departement",y="Marks",hue="Departement",data=df)
plt.show()  # color by category
```

```python
sns.set_theme(style="darkgrid")
```

```python
sns.kdeplot(df["Marks"]) # distribution plot
plt.show()
```
