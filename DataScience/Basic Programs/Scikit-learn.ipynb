```python
import numpy as np
import pandas as pd
df=pd.read_excel("D:\Sumanth.xlsx")
df.head()
```

```python
print(df.columns)
```

```python
from sklearn.model_selection import train_test_split

# Drop rows with missing values in the features or target column
clean_df = df.dropna(subset=["Attendance ", "Marks"])

x = clean_df[["Attendance "]]
y = clean_df[["Marks"]]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression 
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
print(predictions)  # making predictions 
```

```python
new_data=[[90]]  ## predicting the new student data
prediction=model.predict(new_data)
print(prediction)
```

```python
from sklearn.tree import DecisionTreeClassifier ## classification
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```
